```python
df2 = pd.DataFrame()
start = time.time()

date = (datetime.datetime.today()  -  datetime.timedelta(days = 234)).date()
for k in range(31):
    yesterday = (date + datetime.timedelta(days = k)).strftime('%Y-%m-%d')
    url =  "https://backstage.taboola.com/backstage/api/1.0/" + account_id + "/reports/top-campaign-content/dimensions/item_breakdown?start_date=" + yesterday  + "&end_date=" +  yesterday

    headers = {
        "Authorization": "Bearer " + access_token,
        "Content-Type": "application/json"
    }

    # Send the request, create dataframe from results
    response = requests.get(url, headers=headers)
    if len(response.json()['results']) == 0:
        print('nothing')
        continue
    df = pd.DataFrame(response.json()['results'])

    # coerce data types within the dataframe
    df['item'] = pd.to_numeric(df['item'])
    df['campaign'] = pd.to_numeric(df['campaign'])
    df['content_provider'] = pd.to_numeric(df['content_provider'])
    df['impressions'] = pd.to_numeric(df['impressions'])
    df['visible_impressions'] = pd.to_numeric(df['visible_impressions'])
    df['ctr'] = pd.to_numeric(df['ctr'])
    df['vctr'] = pd.to_numeric(df['vctr'])
    df['clicks'] = pd.to_numeric(df['clicks'])
    df['cpc'] = pd.to_numeric(df['cpc'])
    df['cvr'] = pd.to_numeric(df['cvr'])
    df['cvr_clicks'] = pd.to_numeric(df['cvr_clicks'])
    df['cvr_views'] = pd.to_numeric(df['cvr_views'])
    df['cpa_clicks'] = pd.to_numeric(df['cpa_clicks'])
    df['cpa_views'] = pd.to_numeric(df['cpa_views'])
    df['actions'] = pd.to_numeric(df['actions'])
    df['actions_num_from_clicks'] = pd.to_numeric(df['actions_num_from_clicks'])
    df['actions_num_from_views'] = pd.to_numeric(df['actions_num_from_views'])
    df['cpm'] = pd.to_numeric(df['cpm'])
    df['vcpm'] = pd.to_numeric(df['vcpm'])
    df['spent'] = pd.to_numeric(df['spent'])
    df['conversions_value'] = pd.to_numeric(df['conversions_value'])
    df['roas'] = pd.to_numeric(df['roas'])
    df['date'] = yesterday

    df2 = pd.concat([df2, df])
    print(yesterday, time.time() - start)
```
