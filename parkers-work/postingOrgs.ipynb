```python
import requests
import json
import pandas as pd

# Define the URL for your WordPress REST API endpoint
url = "https://allveteran.com/wp-json/wp/v2/orgs"

# Define the headers, including the Authorization header
headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Basic YWxsdmV0ZXJhbjprOEdhIDFzS2Ega3EwZCBYVWFmIHo3SlMgdElRdw==',
    'User-Agent': 'PostmanRuntime/7.41.0'
}

# Loop through each row in the DataFrame and post the data
for index, row in df.iterrows():
    # Create the payload for the current row
    payload = {
        "title": row["orgName"],
        "slug": row["orgName"].lower(),
        "status": "publish",  # You can change this to 'draft', 'pending', 'private', etc.
        "acf": {
            "name": row["orgName"],
            "phone": row["orgPhone"],
            "address": row["address"],
            "website": row["website"],
            "state": row["orgState"],
            "AttorneyList": row["repName"],
            "rating": str(row["rating"]),
            "coordinates": row["coordinates(LatLong)"]
        }
    }

    # Convert the payload to JSON format
    payload_json = json.dumps(payload)

    # Send the POST request to the WordPress REST API
    response = requests.post(url, headers=headers, data=payload_json)

    # Print the response
    print(f"Response for {row['orgName']}: {response.text}")

    # Check if the post was created successfully
    if response.status_code == 201:
        print(f"Successfully created post for {row['orgName']}")
    else:
        print(f"Failed to create post for {row['orgName']}. Status code: {response.status_code}, Response: {response.text}")
```
