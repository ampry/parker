```python
# Convert the 'created_on' column to datetime
df_ourIntake['created_on'] = pd.to_datetime(df_ourIntake['created_on'])

# Define the comparison date as a Timestamp
date_to_compare = pd.to_datetime('2024-01-10 10:05:00.000000')

# Filter the DataFrame based on the converted 'created_on' column
df_ourIntake = df_ourIntake[df_ourIntake['created_on'] >= date_to_compare]
```

```python
# Drop rows with NaN values in the 'Phone' column
df_intakes = df_intakes.dropna(subset=['Phone'])

# Convert phone numbers from floats to integers and remove the leading '1' if present
intakePhoneList = df_intakes['Phone'].apply(lambda x: int(str(int(x))[1:]) if str(int(x)).startswith('1') else int(x)).tolist()
df_intakes['Phone'] = df_intakes['Phone'].apply(lambda x: int(str(int(x))[1:]) if str(int(x)).startswith('1') else int(x))

# Display the updated list
print(intakePhoneList)
```

```python
phoneList = df_medas.Phone.tolist()
```

```python
import re

# Function to safely convert to an integer
def convert_phone_to_int(phone):
    # Remove any non-numeric characters (e.g., hyphens, periods, and plus signs)
    phone_str = re.sub(r'\D', '', phone)

    # Convert to integer
    return int(phone_str)

# Apply conversion to each phone number in the list
phoneList = [convert_phone_to_int(phone) for phone in phoneList]

print(phoneList)
```

```python
# Loop through the phone numbers in df_ourMeda and remove each from phoneList
for phone in df_ourIntake['phone']:
    if phone in intakePhoneList:
        intakePhoneList.remove(phone)  # Remove only one occurrence

# Display the updated phoneList
print(intakePhoneList)
```

```python
# Loop through the phone numbers in df_ourMeda and remove each from phoneList
for phone in df_ourMeda['phone']:
    if phone in phoneList:
        phoneList.remove(phone)  # Remove only one occurrence

# Display the updated phoneList
print(phoneList)
```
